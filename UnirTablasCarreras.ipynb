```python
#Unir todas las tablas de carrera 
import pandas as pd
tabla_carrera20251 = 'tabla_carrera20251.csv'
tabla_carreraA = 'tabla_carreraA.csv'
tabla_carreraBCDE = 'tabla_carreraBCDE.csv'
tabla_carreraESPECIAL = 'tabla_carreraESPECIAL.csv'
tabla_carreraDEPORTISTA = 'tabla_carreraDEPORTISTAS.csv'
df1 = pd.read_csv(tabla_carrera20251)

df2 = pd.read_csv(tabla_carreraA)

df3 = pd.read_csv(tabla_carreraBCDE)

df4 = pd.read_csv(tabla_carreraESPECIAL)

df5 = pd.read_csv(tabla_carreraDEPORTISTA)

concatenar = pd.concat([df1,df2,df3,df4,df5])

# Eliminar duplicados
eliminar = concatenar.drop_duplicates()
eliminar.to_csv('tabla_carreras_unida.csv', index=False)
```

```python
def cambiar_modalidad_especial(id_modalidad):
    if id_modalidad == 0:
        return 6
    elif id_modalidad == 1:
        return 7
    elif id_modalidad == 2:
        return 8
    elif id_modalidad == 3:
        return 9
def cambiar_modalidad_deportista(id_modalidad):
    if id_modalidad == 0:
        return 10
    elif id_modalidad == 1:
        return 11

df4['IdModalidad'] = df4['IdModalidad'].apply(cambiar_modalidad_especial)
df5['IdModalidad'] = df5['IdModalidad'].apply(cambiar_modalidad_deportista)
df4.to_csv('especial.csv', index=False)

df5.to_csv('deportista.csv', index=False)

concatenarfinal = pd.concat([df1,df2,df3,df4,df5])
concatenarfinal.to_csv('tabla_resultados_unidos.csv', index=False)
```

```python
# Unir todas las modalidades de ingreso

tabla_modalidades20251 = 'tabla_modalidad20251.csv'
tabla_modalidadesA = 'tabla_modalidadA.csv'
tabla_modalidadesBCDE = 'tabla_modalidadBCDE.csv'
tabla_modalidadesESPECIAL = 'tabla_modalidadESPECIAL.csv'
tabla_modalidadesDEPORTISTA = 'tabla_modalidadDEPORTISTAS.csv'

df1 = pd.read_csv(tabla_modalidades20251)

df2 = pd.read_csv(tabla_modalidadesA)

df3 = pd.read_csv(tabla_modalidadesBCDE)

df4 = pd.read_csv(tabla_modalidadesESPECIAL)

df5 = pd.read_csv(tabla_modalidadesDEPORTISTA)

concatenarfinal = pd.concat([df1,df2,df3,df4,df5])

concatenarfinal = concatenarfinal.drop_duplicates()

print(concatenarfinal)
```

```python
def cambiar_id_modalidad(modalidad):
    if 'educación básica' in modalidad:
        return 0
    elif 'primeros puestos' in modalidad:
        return 1
    elif 'victimas' in modalidad:
        return 2
    elif 'comunidades' in modalidad:
        return 3
    elif 'personas con discapacidad' in modalidad:
        return 4
    elif 'plan integral' in modalidad:
        return 5
    elif 'traslado interno' in modalidad:
        return 6
    elif 'graduados' in modalidad:
        return 7
    elif 'traslado externo nacional' in modalidad:
        return 8
    elif 'traslado externo internacional' in modalidad:
        return 9
    elif 'deportista calificado' in modalidad:
        return 10
    elif 'deportistas calificados de alto nivel' in modalidad:
        return 11

concatenarfinal['id_modalidad'] = concatenarfinal['Modalidad'].apply(cambiar_id_modalidad)
concatenarfinal.to_csv('tabla_modalidades_unidas.csv', index=False)
```
